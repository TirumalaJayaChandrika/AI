```python
# 8 puzzle
initial = [[1, 2, 3], [0, 4, 6], [7, 5, 8]]
goal = [[1,2,3],[4,5,6],[7,8,0]]
from queue import PriorityQueue
import copy
open_list = PriorityQueue()
path = []
g = 0
def findHeuristic(cur):
  h = 0
  for i in range(len(cur)):
    for j in range(len(cur)):
      if(goal[i][j] != cur[i][j] and cur[i][j] != 0):
        h += 1
  return h+g
def up(cur_p,i,j):
  cur = copy.deepcopy(cur_p)
  temp = cur[i][j]
  cur[i][j]=cur[i-1][j]
  cur[i-1][j] = temp
  open_list.put((findHeuristic(cur),cur))
def down(cur_p,i,j):
  cur = copy.deepcopy(cur_p)
  temp = cur[i][j]
  cur[i][j]=cur[i+1][j]
  cur[i+1][j] = temp
  open_list.put((findHeuristic(cur),cur))
def left(cur_p,i,j):
  cur = copy.deepcopy(cur_p)
  temp = cur[i][j]
  cur[i][j]=cur[i][j-1]
  cur[i][j-1] = temp
  open_list.put((findHeuristic(cur),cur))
def right(cur_p,i,j):
  cur = copy.deepcopy(cur_p)
  temp = cur[i][j]
  cur[i][j]=cur[i][j+1]
  cur[i][j+1] = temp
  open_list.put((findHeuristic(cur),cur))
def findPos(cur):
  global g
  g += 1

  for x in range(len(cur)):
    for y in range(len(cur)):
      if(cur[x][y] == 0):
        i = x
        j = y
        break
  if(i == 0 and j == 0):
    right(cur, i, j)
    down(cur, i, j)
  elif(i == 0 and j == 1):
    right(cur, i, j)
    down(cur, i, j)
    left(cur, i, j)
  elif(i == 0 and j == 2):
    down(cur, i, j)
    left(cur, i, j)
  if(i == 1 and j == 0):
    right(cur, i, j)
    down(cur, i, j)
    up(cur, i, j)
  if(i == 1 and j == 1):
    right(cur, i, j)
    down(cur, i, j)
    left(cur, i, j)
    up(cur, i, j)
  elif(i == 1 and j == 2):
    left(cur, i, j)
    down(cur, i, j)
    up(cur, i, j)
  elif(i == 2 and j == 0):
    left(cur, i, j)
    up(cur, i, j)
  elif(i == 2 and j == 1):
    left(cur, i, j)
    right(cur, i, j)
    up(cur, i, j)
  elif(i == 2 and j == 2):
    left(cur, i, j)
    up(cur, i, j)



def apuzzle(initial):
  open_list.put((0,initial))
  while open_list.empty() == False:
    f_val, cur = open_list.get()
    if(cur == goal):
      print("goal state is reached with h:",f_val)
      print(path)
      break
    path.append(cur)
    findPos(cur)
apuzzle(initial)

```

```python
# ao star

def cost(h, con):
    cost = {}
    if 'AND' in con:
        val = con['AND']
        path_k = ' AND '.join(val)
        path_c = sum([1+h[i] for i in val])
        cost[path_k] = path_c
    if 'OR' in con:
        val = con['OR']
        path_k = ' OR '.join(val)
        path_c = min([1+h[i] for i in val])
        cost[path_k] = path_c
#     print(con, cost)
    return cost
def updatecost(h, conditions):
    main_nodes = list(conditions.keys())
    main_nodes.reverse()
    d = {}
    for key in main_nodes:
        con = conditions[key]
        c = cost(h, con)
        print(key," : ",con,">>>",c)
        h[key] = min(c.values())
        d[key] = cost(h, con)
#     print(d)
    return d
        
def shortestPath(start,cost, h):
    path = start
    if start in list(cost.keys()):
        min_cost = min(cost[start].values())
        key = list(cost[start].keys())
        val = list(cost[start].values())
        ind = val.index(min_cost)
        nex = key[ind].split()
        if(len(nex) == 1):
            start = nex[0]
            path += "<---"+shortestPath(start,cost, h)
        else:
            path += "<---("+key[ind]+")"
            start = nex[0]
            path += "["+shortestPath(start,cost, h)+"+"
            
            start = nex[-1]
            path += shortestPath(start,cost, h)+"]"
    return path
            
h = {'A': -1, 'B': 5, 'C': 2, 'D': 4, 'E': 7, 'F': 9, 'G': 3, 'H': 0, 'I':0, 'J':0}

conditions = {
'A': {'OR': ['B'], 'AND': ['C', 'D']},
'B': {'OR': ['E', 'F']},
'C': {'OR': ['G'], 'AND': ['H', 'I']},
'D': {'OR': ['J']}
}
print(shortestPath('A',updatecost(h, conditions),h))
```

```python
# min max
def minmax(curDepth, maxTurn, scores, treeDepth):
    if curDepth == treeDepth:
        if maxTurn:
            return max(scores)
        else:
            return min(scores)
    
    if(maxTurn):
        curMax = float('-inf')
        for i in scores:
            curMax = max(curMax, minmax(curDepth+1, False, i, treeDepth))
        return curMax
    else:
        curMin = float('inf')
        for i in scores:
            curMin = min(curMin, minmax(curDepth+1, True, i, treeDepth))
        return curMin
scores = [[[2,3],[5,9]],[[0,1],[7,5]]]
treeDepth = 2
print("optimal val:",minmax(0, True, scores, treeDepth))
        
```
